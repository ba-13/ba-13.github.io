```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
```

```python
FIGSIZE = (15,5)
```

```python
f = 2
```

```python
x = np.linspace(0, 4*np.pi, 1000)
cos = np.cos(f*x)
```

```python
fig, ax = plt.subplots(1, 1, figsize=FIGSIZE)
ax.plot(x, cos)
ax.axvline(x=0, c="white", label="x=0")
ax.axhline(y=0, c="white", label="y=0")
ax.axvline(x=np.pi, c="yellow", label="x=π")
ax.set_title(f"cos({f}x)")
ax.legend()
```

```python
f1 = 1
f2 = 2
```

```python
x = np.linspace(0, 4, 1000)
cos1 = np.cos(2*np.pi*f1*x)
cos2 = np.cos(2*np.pi*f2*x)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15,8))
ax[0].plot(x, cos1)
ax[0].axvline(x=0, c="white", label="x=0")
ax[0].axhline(y=0, c="white", label="y=0")
ax[0].axvline(x=np.pi, c="yellow", label="x=π")
ax[0].legend()
ax[0].set_title(f"cos(2π{f1}*t)")

ax[1].plot(x, cos2)
ax[1].axvline(x=0, c="white", label="x=0")
ax[1].axhline(y=0, c="white", label="y=0")
ax[1].axvline(x=np.pi, c="yellow", label="x=π")
ax[1].legend()
ax[1].set_title(f"cos(2π{f2}*t)")
```

```python
x = np.linspace(0, 10, 1000)
amplitudes = [[5.535120203969017, 8.379075027996663, 2.4179404450317454, 8.80427381827397], [4.204233843140538, 8.809556566145538, 6.8419681219436885, 1.4146715362057871]]
func = np.zeros(x.shape)
freq = len(amplitudes[0])
functions = []
# amplitudes = [[],[]]
for f in range(0, freq):
    # camplitude = 10*np.random.random()
    # samplitude = 10*np.random.random()
    # amplitudes[0].append(camplitude)
    # amplitudes[1].append(samplitude)
    camplitude = amplitudes[0][f]
    samplitude = amplitudes[1][f]
    cos = camplitude * np.cos(2*np.pi*0.3*f*x)
    sin = samplitude * np.sin(2*np.pi*0.3*f*x)
    func += cos + sin
    functions.append(func.copy())
print(amplitudes)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15,8))
ax[0].plot(x[:700], func[:700])
ax[0].axvline(x=0, c="white", label="x=0")
ax[0].axhline(y=0, c="white", label="y=0")
ax[0].axvline(x=np.pi, c="yellow", label="x=π")
ax[0].legend()
ax[0].set_title(f"random function")

ax[1].plot(x, func)
ax[1].axvline(x=0, c="white", label="x=0")
ax[1].axhline(y=0, c="white", label="y=0")
ax[1].axvline(x=np.pi, c="yellow", label="x=π")
ax[1].legend()
ax[1].set_title(f"showing periodicity of above function")
```

```python
fig, ax = plt.subplots(4, 1, figsize=(15, 12))
for i in range(freq):
    ax[i].plot(x, functions[i])
    ax[i].axvline(x=0, c="white", label="x=0")
    ax[i].axhline(y=0, c="white", label="y=0")
    ax[i].set_ybound(-15, 30)
    ax[i].axvline(x=np.pi, c="yellow", label="x=π")
    ax[i].set_title(f"{i} frequency component")
    # ax[i].legend()
fig.tight_layout()
```
